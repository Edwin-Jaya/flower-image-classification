```python
def counting_images_files_in_folder(dataset_path)->None:
    image_ext = [".jpg",".png","svg",".jpeg"]
    count=0
    for i in os.listdir(dataset_path):
        extension = os.path.splitext(i)[-1]
        if extension in image_ext: 
            count+=1
    print("Images files in folder {} : {} files".format(dataset_path,count))
```

```python
def creating_folder(path)->None:
    isExists=os.path.exists(path)
    if not isExists:
        os.mkdir(path)
        print("Folder successfully created!")
    else:
        print("Folder already exists!")
```

```python
def seperate_image(path,flowers_type):
    flowers_arr=[]
    for i in os.listdir(path):
        flowers_name = os.path.splitext(i)[0]
        if flowers_type in flowers_name:
            flowers_arr.append(i)
    return flowers_arr
```

```python
tulips_image = seperate_image(dataset_path,"tulip")
orchid_image = seperate_image(dataset_path,"orchids")
peonies_image = seperate_image(dataset_path,"peonies")
hydra_image = seperate_image(dataset_path,"hydrangeas")
lilies_image = seperate_image(dataset_path,"lilies")
gardenias_image = seperate_image(dataset_path,"gardenias")
garden_roses_image = seperate_image(dataset_path,"garden_roses")
daises_image = seperate_image(dataset_path,"daisies")
hibiscus_image = seperate_image(dataset_path,"hibiscus")
bougainvillea_image = seperate_image(dataset_path,"bougainvillea")
```

```python
def create_data(data,ratio)->int:
    total_data = (len(data)*ratio)
    return total_data
```

```python
def shuffling_data(data)->None:
    random.shuffle(data)
```

```python
shuffling_data(tulips_image)
shuffling_data(orchid_image)
shuffling_data(peonies_image)
shuffling_data(hydra_image)
shuffling_data(lilies_image)
shuffling_data(gardenias_image)
shuffling_data(garden_roses_image)
shuffling_data(daises_image)
shuffling_data(hibiscus_image)
shuffling_data(bougainvillea_image)
```

```python
def moving_image(data, source, train_size, val_size, train_path, validation_path, label)->None:
    for i, file in enumerate(data):
        if i<train_size:
            dest_folder = train_path+label
        elif i<train_size+val_size:
            dest_folder = validation_path+label
        isExists=os.path.exists(dest_folder)
        if not isExists:
            os.mkdir(dest_folder)
        shutil.move(os.path.join(source,file),os.path.join(dest_folder,file))
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 40,
    zoom_range = 0.2,
    horizontal_flip= True,
    vertical_flip=True,
    fill_mode = "nearest",
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2
)
val_datagen = ImageDataGenerator(
    rescale = 1./255
)
```

```python
train_gen = train_datagen.flow_from_directory(
    train_path,
    batch_size=32,
    target_size=(150,150),
    class_mode="categorical"
)
```

```python
test_gen = train_datagen.flow_from_directory(
    validation_path,
    batch_size=32,
    target_size=(150,150),
    class_mode="categorical"
)
```

```python
def build_model(model_filter,model_layer):
    model = tensorflow.keras.models.Sequential([
    tensorflow.keras.layers.Conv2D(model_filter, model_layer, padding="same", activation='relu', input_shape=(150, 150, 3)),
    tensorflow.keras.layers.MaxPooling2D(2, 2),
    tensorflow.keras.layers.Flatten(),
    tensorflow.keras.layers.Dense(10, activation='softmax')
    ])
    return model
```

# Model 1

```python
model = build_model(64,(5,5))
```

```python
model.summary()
```

```python
def build_compiler(model,loss,optimizer)->None:
    model.compile(loss=loss,
                  optimizer=optimizer,
                  metrics=['accuracy'])
```

```python
optimizer = tensorflow.optimizers.SGD(learning_rate=0.01, momentum=0.9)
build_compiler(model,"categorical_crossentropy",optimizer)
```

```python
def create_plt(history)->None:
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
acc_treshold = 0.85
```

```python
class myCallback(tensorflow.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy') > acc_treshold):
          print("\nReached %2.2f%% accuracy, so stopping training!!" %(acc_treshold*100))
          self.model.stop_training = True
```

```python
callbacks = myCallback()
```

```python
history = model.fit(train_gen,
                    validation_data=test_gen,
                    batch_size=32,
                    epochs=15,
                    verbose=1,
                    callbacks=[callbacks])
```

```python
create_plt(history)
```
